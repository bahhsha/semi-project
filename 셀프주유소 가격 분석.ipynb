```python
# 서울시 구별 주유소 가격 정보 가져오기
# URL : https://www.opinet.co.kr/searRgSelect.do
driver = webdriver.Chrome('../driver/chromedriver.exe')
driver.get('https://www.opinet.co.kr/searRgSelect.do')
```

```python
# //*[@id="SIGUNGU_NM0"]
# 구가 들어있는 select box 선택
gu_list_raw = driver.find_element(By.XPATH, '//*[@id="SIGUNGU_NM0"]')
print(gu_list_raw)
# ctrl+shift+'-' 셀을 쪼개줌
```

```python
gu_list = gu_list_raw.find_elements(By.TAG_NAME,'option')
gu_list
```

```python
gu_names = [option.get_attribute('value') for option in gu_list]
gu_names
```

```python
gu_names.remove('')
gu_names
```

```python
# 중랑구가 선택되어있는지 확인 테스트
# 구가 들어있는 select Box 선택
elem_gu = driver.find_element(By.ID,'SIGUNGU_NM0')
# 중랑구 선택
elem_gu.send_keys(gu_names[-1])
# 변경 확인
```

```python
# 조회 버튼 클릭
# 조회 버튼 선택 => XPATH, # 클릭
xpath = '//*[@id="searRgSelect"]'
elem_search = driver.find_element(By.XPATH, xpath).click()
# 클릭
```

```python
# 엑셀 저장 버튼 클릭
# 버튼 선택, 클릭
xpath = '//*[@id="glopopd_excel"]'
driver.find_element(By.XPATH,xpath).click()
```

```python
gu_names
```

```python
# 전체 구 데이터 추출
import time
from tqdm import tqdm

for gu in tqdm(gu_names):
    elem = driver.find_element(By.ID, 'SIGUNGU_NM0')
    elem.send_keys(gu) # 검색하고자 하는 구 이름 서버로 전송
    time.sleep(2) # 2초간 결과를 기다린다.
    
    # 조회 버튼 클릭
    xpath = '//*[@id="searRgSelect"]'
    elem_search = driver.find_element(By.XPATH, xpath).click()
    time.sleep(1)
    
    # 엑셀 저장 버튼 클릭
    xpath = '//*[@id="glopopd_excel"]'
    driver.find_element(By.XPATH,xpath).click()
    time.sleep(1)
    
```

```python
# 2018년 데이터 
stations_files = glob('../data/drive-download-20220930T010652Z-001/지역*.xls')
tmp_raw = []
for file_name in stations_files:
    tmp_df = pd.read_excel(file_name, header=2)
    tmp_raw.append(tmp_df)    
station_raw = pd.concat(tmp_raw)
station_raw.info()
```

```python
# 데이터 정리
# 상호, 주소, 휘발유, 셀프여부, 상표
stations_df = pd.DataFrame({
      '상호' : station_raw['상호'] # Series
    , '주소' : station_raw['주소']
    , '가격' : station_raw['휘발유']
    , '셀프' : station_raw['셀프여부']
    , '상표' : station_raw['상표']
})
stations_df.head()
```

```python
# 주소에서 구 추출
stations_df['구'] = [addr.split()[1] for addr in stations_df['주소']]
stations_df['구']
```

```python
# '-' : 현재 가격을 입력하지 않은 주유소 확인
stations_df[stations_df['가격']=='-']
```

```python
# 가격이 없는 주유소 제외하고 새로운 데이터프레임을 생성 => boolean indexing 사용
stations_df = stations_df[stations_df['가격'] != '-']
```

```python
stations_df[stations_df['가격']=='-'] # 가격이 없는 주유소를 제외해서 결과값이 나오지 않는다
```

```python
# 가격을 숫자 (실수) => astype()
stations_df['가격'] = [float(price) for price in stations_df['가격']]
```

```python
# 상표 별, 셀프 가격 분석 => boxplot으로 구현해보자!
plt.figure(figsize=(12,8))
sns.boxplot(
      data = stations_df
    , x = '상표'
    , y = '가격'
    , hue = '셀프'
)
plt.show()
# sk에너지가 가장 비싸다! 알뜰 주유소가 가장 쌈..
```

```python
# 상표 별, 셀프 가격 분석 => boxplot으로 구현해보자!
plt.figure(figsize=(8,6))
sns.boxplot(
      data = stations_df
    , x = '상표'
    , y = '가격'
    , hue = '셀프'
    , palette='Set1' # 색을 palette로 줄 수 있다
)
plt.show()
 
```

```python
# 구별 가격 평균
pd.pivot_table( # pivot과 pivot_table의 차이 비교해보기
      data=stations_df
    , index=['구', '상표', '셀프'] # DataFrame임
    , values=['가격']
    , aggfunc=np.mean
)
```
